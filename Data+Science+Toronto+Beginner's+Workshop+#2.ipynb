Pandas has the loc function (Short for location) which is similar to the calling function we saw with arrays. The difference here is that since Arrays were one dimensional you could simply use A[0] or A[2:50] to call elements in the array. Since we are working with data frames, we have 2 dimensions (rows AND columns) to deal with when calling elements in a data frame. When using the loc function we first call the rows (or index) we want to use then the columns. Notice the differences in syntax used between the 2; when calling rows, I can slice (give a beginning and ending column) whereas for columns, I call each 1 by 1.

```python
df.loc[:,'D']
```

If we want to call an entire column, we can use : to call all rows, and pick out which column we'd like. Keep in mind that using the location function produces a new data frame out of the values you called.

Like arrays, we can pass boolean statements over entries in our dataframe. Here, I tried calling all elements in the dataframe A that were greater than 0. Entries that fulfilled this requirement have their value shown while elements that didnt have NaN (or Not A Number) in their place. NaN is used as a mask value in such a case. Mask values are useful as they help with the data cleaning process by showing which data points are not needed.

```python
A[A.A>0]
```

We can also pass boolean statements over just columns as well. Above, we called all rows where the elements in the column A, in our data frame A, had a value greater than 0.
