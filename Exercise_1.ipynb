```
x = np.linspace(1,25,25)
print (x)
```

Above I used the linspace function to generate 25 evenly spaced numbers between 1 and 25. 

The linspace function takes 3 arguments in this order: the starting number, the ending number, number of points  

NOTE: When I say evenly spaced it means that a number is the same difference away from the number before and after it. Ex. In the sequence, (3,4,5) the number 4 is greater than 3 (the previous number) by 1 and less than 5 (the following number) by 1.
